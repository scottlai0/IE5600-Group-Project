```python
import pandas as pd 
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
start = datetime(2019,12,30)
end = datetime(2020,9,6)

eps = [0]*10
company = [0]*10
```

```python
facebook = yf.download('FB', start=start, progress = False)
amazon = yf.download('AMZN', start=start, end=end,progress=False)
google = yf.download('GOOG', start=start, end=end,progress=False)
apple = yf.download('AAPL',start=start,end=end,progress=False)
netflix = yf.download('NFLX',start=start, end=end,progress=False)
norvatis = yf.download('NVS', start=start, end=end,progress=False)
abbott = yf.download('ABT', start=start, end=end,progress=False)
pfizer = yf.download('PFE', start=start, end=end,progress=False)
merck = yf.download('MRK', start=start, end=end,progress=False)
jnj = yf.download('JNJ', start=start, end=end,progress=False)

facebook_close = facebook['Adj Close']
amazon_close = amazon['Adj Close']
google_close = google['Adj Close']
apple_close = apple['Adj Close']
netflix_close = netflix['Adj Close']
norvatis_close = norvatis['Adj Close']
abbott_close = abbott['Adj Close']
pfizer_close = pfizer['Adj Close']
merck_close = merck['Adj Close']
jnj_close = jnj['Adj Close']
```

```python
facebook_eps = yf.Ticker('FB').info.get("trailingEps")
amazon_eps = yf.Ticker('AMZN').info.get("trailingEps")
google_eps = yf.Ticker('GOOG').info.get("trailingEps")
apple_eps = yf.Ticker('AAPL').info.get("trailingEps")
netflix_eps = yf.Ticker('NFLX').info.get("trailingEps")
norvatis_eps = yf.Ticker('NVS').info.get("trailingEps")
abbott_eps = yf.Ticker('ABT').info.get("trailingEps")
pfizer_eps = yf.Ticker('PFE').info.get("trailingEps")
merck_eps = 4.10 #yf.Ticker('MRK').info.get("trailingEps")
jnj_eps = yf.Ticker('JNJ').info.get("trailingEps")
```

```python
plt.figure(figsize=(15,8))
plt.plot(facebook_close/facebook_eps/25, 'teal',label = 'Facebook')
plt.plot(amazon_close/amazon_eps/71, 'dodgerblue', label = 'Amazon')
plt.plot(google_close/google_eps/29.5, 'skyblue', label ='Google')
plt.plot(apple_close/apple_eps/22, 'darkturquoise', label ='Apple')
plt.plot(netflix_close/netflix_eps/55, 'royalblue', label ='Netflix')

plt.plot(norvatis_close/norvatis_eps/29, 'lightcoral',label = 'Norvatis')
plt.plot(abbott_close/abbott_eps/49, 'firebrick', label = 'Abbott')
plt.plot(pfizer_close/pfizer_eps/15, 'orangered', label ='Pfizer')
plt.plot(merck_close/merck_eps/22.25, 'tomato', label ='Merck')
plt.plot(jnj_close/jnj_eps/25, 'red', label ='JNJ')

plt.xlabel('Date')
plt.ylabel('PE ratio growth')
plt.legend()

plt.show()
```
