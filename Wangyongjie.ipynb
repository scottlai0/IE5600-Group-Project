```python
apple_fr=requests.get('https://financialmodelingprep.com/api/v3/ratios/AAPL?apikey=ff76fe869a5bf1f03c6f38a0c366980e')
apple_fr=apple_fr.json()
y=[]
x=[]
for i in range(len(apple_fr)):
    y.append(apple_fr[i]['currentRatio'])
    x.append(apple_fr[i]['date'])
figsize=8,5
figure, ax = plt.subplots(figsize=figsize)
for i in ax.get_xticklabels():
    i.set_visible(False)
for i in ax.get_xticklabels()[::20]:
    i.set_visible(True)
plt.plot(x[::-1],y[::-1],'y',label='A',linewidth=2.0,ms=10)
```
